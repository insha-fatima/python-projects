```python
import random
print("Guess the number correctly from 1 to 10,with large value or small value or correct value")
def computer_guess():
  small = 1
  large = 20
  feedback = ''
  while feedback != 'correct':
    if small != large:
      guess = random.randint(1,10)
    feedback = input(f"{guess} the number is large value or small value or correct value: ")
    if feedback == 'large':
      large = guess-1
    elif feedback == 'small':
      small = guess+1
computer_guess()
print("Your guessed number is correct!")
```
