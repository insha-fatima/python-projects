# ***3- FAHRENHEIT INTO CELSIUS***

---



```python
def main():
    a: str = 80
    b: str = (a-32)*5/9
    print("80 degree_Fahrenheit is equal to " + str(b) + " degree_Celsius.")
    x: str = 9.8
    y: str = (x-32)*5/9
    print("9.8 degree_Fahrenheit is equal to " + str(y) + " degree_Celsius.")



# This provided line is required at the end of
# Python file to call the main() function.
if __name__ == '__main__':
    main()
```

# ***4- HOW OLD ARE THEY ?***

---



```python
def main():
    anthon: int = 21
    beth: int = anthon + 6
    chen: int = beth + 20
    drew: int = chen + anthon
    ethan: int = chen
    print("Anthon is " + str(anthon) + " years old.")
    print("Beth is 6 yeras older than Anthon = " + str(beth) + " years old.")
    print("Chen is 20 years older than Beth = " + str(chen) + " years old.")
    print("Drew is old as both Chen and Anthon = " + str(drew) + " years old.")
    print("Ethan is same as Chen age = " + str(ethan) + " years old.")


# This provided line is required at the end of
# Python file to call the main() function.
if __name__ == '__main__':
    main()
```
