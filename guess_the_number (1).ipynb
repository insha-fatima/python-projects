```python
import random

def computer_guess():
  small = 1
  large = 10
  feedback = ''
  print("Guess the number correctly from 1 to 10,with large value or small value or correct value")
  while feedback != 'correct':
    if small != large:
      guess = random.randint(small, large)
    feedback = input(f"{guess} the number is large value or small value or correct value: ").lower()
    if feedback == 'large':
      large = guess-1
    elif feedback == 'small':
      small = guess+1
    elif feedback != 'correct':
      print("Invalid input! Please type 'large', 'small', or 'correct'.")

  print("Your guessed number is correct!")
computer_guess()

```
