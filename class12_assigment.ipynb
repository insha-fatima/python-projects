recursive function for factorial


```python
def factorial(n: int, num: int):
  if n <= 0:
     print("Fictorial of", num,"is:")
     return 1
  else:
    print("num",n)
    return n * factorial(n-1, num)
result = factorial(6, 6)
print(result)




```
