# ***TABLE OF "2" BY FOR LOOP***

---
Here, the table of "2" by FOR LOOP method , in which we can easily make a table of 2 by starting point and increment.


```python
num = range(1, 11)
for num in num:
  print(f"2 * {num} = {2 * num}")
```
